```python
import ipywidgets as ipw
import pickle
import webbrowser
%run APIfunctions.ipynb

with open('loadout_options.data', 'rb') as filehandle:
    loadout_options = pickle.load(filehandle)
    
with open('requirements_tasks.data', 'rb') as filehandle:
    requirements = pickle.load(filehandle)
    
output = ipw.Output()
```

```python
class Loadout:
    def __init__(self, 
                 activity_type = 'None', 
                 activity = 'None', 
                 c_class = 'None', 
                 subclass = 'None', 
                 tree = 'None', 
                 grenade = 'None', 
                 kinetic_weapon = 'None', 
                 energy_weapon = 'None', 
                 energy_type = 'None', 
                 power_weapon = 'None', 
                 power_type = 'None', 
                 armor = 'None'):
        
        self.activity_type = activity_type
        self.activity = activity
        self.c_class = c_class
        self.subclass = subclass
        self.tree = tree
        self.grenade = grenade
        self.kinetic_weapon = kinetic_weapon
        self.energy_weapon = energy_weapon
        self.energy_type = energy_type
        self.power_weapon = power_weapon
        self.power_type = power_type
        self.armor = armor
        
current_loadout = Loadout()     
```

```python
def build_loadout_panel():
    global current_loadout
    global layout_dropdown_inactive
    global layout_dropdown_active
    
    #Dropdown layouts ---------------------------
    
    layout_dropdown_inactive = ipw.Layout(
        width='98%',
        border = '')

    layout_dropdown_active = ipw.Layout(
        width='98%',
        border = '2px solid #00bcd4')
    
    #Activity VBox -------------------------------
    
    label_activity_options = ipw.HTML(
        value = "<b>Activity Options</b>")

    
    dropdown_activity_type = ipw.Dropdown(
        options = loadout_options['Activity Type'],
        layout = layout_dropdown_inactive)
    current_loadout.activity_type = dropdown_activity_type

    dropdown_activity = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.activity = dropdown_activity

    vbox_activity = ipw.VBox(
        children=[
            label_activity_options,
            dropdown_activity_type,
            dropdown_activity])
    
    #Class VBox -----------------------------------=
    
    label_class_options = ipw.HTML(
        value = "<b>Class Options</b>")

    label_class = ipw.Label(
        value='Class:')
    
    dropdown_class = ipw.Dropdown(
        options = loadout_options['Class'],
        layout = layout_dropdown_inactive)
    current_loadout.c_class = dropdown_class

    label_subclass = ipw.Label(
        value='Subclass:')
    
    dropdown_subclass = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.subclass = dropdown_subclass

    label_tree = ipw.Label(
        value='Tree:')
    
    dropdown_tree = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.tree = dropdown_tree

    label_grenade = ipw.Label(
        value='Grenade:')
    
    dropdown_grenade = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.grenade = dropdown_grenade

    vbox_class_options = ipw.VBox(
        children = [
            label_class_options,
            label_class,
            dropdown_class,
            label_subclass,
            dropdown_subclass,
            label_tree,
            dropdown_tree,
            label_grenade,
            dropdown_grenade])
    
    
    #Gear VBox -----------------------------------------
    
    label_gear_options = ipw.HTML(
        value = "<b>Gear Options</b>")

    label_kinetic = ipw.Label(
        value='Kinetic Weapon:')
    
    dropdown_kinetic = ipw.Dropdown(
        options = loadout_options['Kinetic Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.kinetic_weapon = dropdown_kinetic

    label_energy = ipw.Label(
        value='Energy Weapon:')
    
    dropdown_energy = ipw.Dropdown(
        options = loadout_options['Energy Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.energy_weapon = dropdown_energy
    
    dropdown_energy_type = ipw.Dropdown(
        options = loadout_options['Energy Weapon Type'],
        layout = layout_dropdown_inactive)
    current_loadout.energy_type = dropdown_energy_type

    label_power = ipw.Label(
        value='Power Weapon:')
    
    dropdown_power = ipw.Dropdown(
        options = loadout_options['Power Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.power_weapon = dropdown_power
    
    dropdown_power_type = ipw.Dropdown(
        options = loadout_options['Power Weapon Type'],
        layout = layout_dropdown_inactive)
    current_loadout.power_type = dropdown_power_type

    label_armor = ipw.Label(
        value='Armor Set:')
    
    dropdown_armor = ipw.Dropdown(
        options = loadout_options['Armor'],
        layout = layout_dropdown_inactive)
    current_loadout.armor = dropdown_armor
    

    vbox_gear = ipw.VBox(
        children = [
            label_gear_options,
            label_kinetic,
            dropdown_kinetic,
            label_energy,
            dropdown_energy,
            dropdown_energy_type,
            label_power,
            dropdown_power,
            dropdown_power_type,
            label_armor,
            dropdown_armor])
    
    #Loadout Panel -------------------------------
    
    box_spacer = ipw.Box(
        layout=ipw.Layout(
            height='20px'))
    
    vbox_options = ipw.VBox(
        children=[
            vbox_activity,
            box_spacer,
            vbox_class_options,
            box_spacer,
            vbox_gear],
        layout = ipw.Layout(
            border ='1px solid white',
            width ='15%',
            min_width = '250px',
            padding ='10px'))


    #Interact Functions ----------------------------

    def change_class(change):
        if change['new'] == 'Any':
            dropdown_subclass.options = ['Any']
            dropdown_subclass.disabled = True
        else:
            dropdown_subclass.options = loadout_options['Subclass'][change['new']]
            dropdown_subclass.disabled = False

    def change_subclass(change):
        if change['new'] == 'Any':
            dropdown_tree.options = ['Any']
            dropdown_tree.disabled = True
            dropdown_grenade.options = ['Any']
            dropdown_grenade.disabled = True
        else:
            dropdown_tree.options = loadout_options['Tree'][dropdown_class.value][change['new']]
            dropdown_tree.disabled = False
            dropdown_grenade.options = loadout_options['Grenade'][dropdown_class.value][change['new']]
            dropdown_grenade.disabled = False

    def change_activity_type(change):
        if change['new'] == 'Any':
            dropdown_activity.options = ['Any']
            dropdown_activity.disabled = True
        else:
            dropdown_activity.options = loadout_options['Activity'][change['new']]
            dropdown_activity.disabled = False

    dropdown_class.observe(change_class, names = 'value')
    dropdown_subclass.observe(change_subclass, names = 'value')
    dropdown_activity_type.observe(change_activity_type, names = 'value')
    
    return vbox_options
```

```python
class Task:
    def __init__(self, task, path, name, description, objectives):
        self.task = task
        self.path = path,
        self.name = name,
        self.description = description,
        self.objectives = objectives
        
        
```

```python
def build_objective():
        
    button_objective = ipw.Button(
        layout = ipw.Layout(
            width = '32px',
            height = '32px',
            order = '-2'))

    intProgress_objective = ipw.IntProgress(
        value=30,
        min=0,
        max=100,
        step=1,
        layout = ipw.Layout(
            width = '100%',
            height = '32px',
            overflow = 'visible'))

    label_objective = ipw.Label(
         value = f'\xa0\xa0\xa0objective', 
         layout = ipw.Layout(
            padding='1px',
            width = '0px',
            overflow = 'visible',
            order='-1'))

    button_edit = ipw.Button(
        description ='EDIT', 
        layout = ipw.Layout(
            width='60px',
            display = 'none'
        )
    )
    

    hbox_objective = ipw.HBox(
        children = [
            button_objective,
            intProgress_objective, 
            label_objective, 
            button_edit])
        
    def toggle_objective(b):
        global active_objectives
        if b.objective_hash in active_objectives:
            active_objectives.remove(b.objective_hash)
        else:
            active_objectives.append(b.objective_hash)
        refresh_loadout()
        refresh_task_view()
    button_objective.on_click(toggle_objective)
    
    
    def edit_objective(b):
        dropdown_triumph_req.value = b.task_hash
        dropdown_objective_req.value = b.objective_hash
    button_edit.on_click(edit_objective)
        
    return hbox_objective
```

```python
task_widgets = []

def build_task_widget():
    global task_widgets

    label_path = ipw.HTML(
        value = '<font size="1">Path/Path/Path/Path/</font size>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'))
    
    label_name = ipw.HTML(
        value = '<b>Task Name</b>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'))
        
    label_description = ipw.HTML(
        value = '<i>test test test test test test test test test test test test test test test test test test test test test test test test </i>',
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap'))
        
    vbox_objectives = ipw.VBox(
        children = [build_objective() for i in range(6)])

    vbox_triumph = ipw.VBox(
        children = [
            label_path,
            label_name,
            label_description, 
            vbox_objectives],
        layout = ipw.Layout(
            margin = '10px',
            padding='10px',
            border='3px solid white',
            flex='1 1 0%', 
            width='auto',
            min_width='400px',
            overflow = 'visible'))
    
    task = Task(vbox_triumph, label_path, label_name, label_description, vbox_objectives)
    task_widgets.append(task)

    return vbox_triumph
```

```python
def build_task_panel():
    
    box_tasks = ipw.Box(
        children = [build_task_widget() for i in range(15)], 
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap',
            align_items='stretch'))
    
    return box_tasks
```

```python
class Filter:
    def __init__(self, show = 'None', compelete = 'None', conflicting = 'None', label = 'None', next = 'None', previous = 'None', refresh = 'None'):
        
        self.show = show
        self.compelete = compelete
        self.conflicting = conflicting
        self.label = label
        self.next = next
        self.previous = previous
        self.refresh = refresh
        
task_filter = Filter()  
```

```python
def build_filter_panel():
    
    checkbox_complete = ipw.Checkbox(
        description='Hide Complete Objectives'
    )
    task_filter.complete = checkbox_complete
    
    checkbox_conflicting = ipw.Checkbox(
        description='Hide Conflicting Objectives'
    )
    task_filter.conflicting = checkbox_conflicting
    
    dropdown_show = ipw.Dropdown(
        description = "Show:",
        options= [
            "All",
            "Active Objectives",
            "Pursuits",
            "Milestones",
            "Triumphs",
            "Collections"
        ]
    )
    task_filter.show = dropdown_show
    
    vbox_filters = ipw.VBox(
        children = [
            dropdown_show,
            checkbox_complete,
            checkbox_conflicting])
    
    label_current_page = ipw.HTML(
        value = 'Showing __-__ of ___')
    task_filter.label = label_current_page
    
    button_previous_page = ipw.Button(
        description = 'Previous',
        layout = ipw.Layout(
            width='100px'))
    task_filter.previous = button_previous_page
    
    button_next_page = ipw.Button(
        description = 'Next',
        layout = ipw.Layout(
            width='100px'))
    task_filter.next = button_next_page
    
    hbox_change_page = ipw.HBox(
        children = [
            button_previous_page,
            button_next_page])
    
    button_refresh = ipw.Button(
        description='Refresh',
        layout = ipw.Layout(
            width='186px'))
    task_filter.refresj = button_refresh
    
    vbox_page = ipw.VBox(
        children=[
            label_current_page,
            hbox_change_page,
            button_refresh],
        layout = ipw.Layout(
            width='200px',
            padding = '5px',
            justify_items = 'flex-end'))
    
    hbox_filter_panel = ipw.HBox(
        children = [
            vbox_filters,
            vbox_page],
        layout = ipw.Layout(
            justify_content = 'space-between',
            height = '300px'))
    
    def previous_page(b):
        global current_page    
        current_page -= 1
        refresh_task_view()


    def next_page(b):
        global current_page
        current_page += 1
        refresh_task_view()

    button_previous_page.on_click(previous_page)
    button_next_page.on_click(next_page)


    def refresh_button(b):
        filter_tasks()
        refresh_task_view()

    button_refresh.on_click(refresh_button)
    
    
    return hbox_filter_panel
```

```python
def interface_gamertag_search():
    text_gamertag = ipw.Text(
        description = 'Gamertag:'
    )

    button_gamertag = ipw.Button(
        description = 'Search'
    )

    hbox_gamertag = ipw.HBox(
        children = [
            text_gamertag,
            button_gamertag
        ]
    )

    def search_gamertag(b):
        success = get_membershipId(text_gamertag.value)
        if success:
            output.clear_output()
            with output: display(authorize())
        else:
            with output: print(f'Unable to find {gamertag}!')
        
    button_gamertag.on_click(search_gamertag)
    
    return hbox_gamertag
```

```python
def authorize():

    label_authenticate =  ipw.Label(
        value = 'In order to authenticate, sign in with Bungie to get an authentication code'
    )

    button_sign_in = ipw.Button(
        description = 'Sign In'
    )

    text_code = ipw.Text(
        description="Code:"
    )

    button_submit_code = ipw.Button(
        description='Submit'
    )

    hbox_code = ipw.HBox(
        children = [
            text_code,
            button_submit_code
        ]
    )
    vbox_authenticate = ipw.VBox(
        children = [
            label_authenticate,
            button_sign_in,
            hbox_code
        ]
    )

    def sign_in(b):
        webbrowser.open(f'https://www.bungie.net/en/OAuth/Authorize?client_id={clientId}&response_type=code', new=1)
    button_sign_in.on_click(sign_in)

    def submit_code(b):
        success = authenticate(text_code.value)
        if success:
            output.clear_output()
            loading = start_loading()
            with output: display(loading)
            load_app(loading)
        else:
            output.clear_output()
            with output: display(authorize())
            with output: print("Please Try Again")

    button_submit_code.on_click(submit_code)
    
    return vbox_authenticate
```

```python
def initialize_interface():
    
    task_panel = ipw.VBox(
        children = [
            build_filter_panel(),
            build_task_panel()
        ]
    )
    
    interface = ipw.HBox(
        children = [
            build_loadout_panel(),
            task_panel
        ],
        layout = ipw.Layout(
            height = '800px'
        )
    )

    return interface
```

```python
def start_loading():
    floatProgress_loading = ipw.FloatProgress(
        description = "Loading:",
        min = 0,
        max = 6,
        value = 0
    )
    
    label_loading = ipw.Label(value = 'Initializing...')
    
    hbox_loading = ipw.HBox(
        children = [
            floatProgress_loading,
            label_loading
        ]
    )
    
    return hbox_loading
    
```

```python
def load_app(loading):
    global interface
    
    progressbar = loading.children[0]
    label = loading.children[1]
    progressbar.value = 1
    label.value = 'Building interface...'
    interface = initialize_interface()
    progressbar.value = 2
    label.value = 'Getting profile data (characters)...'
    get_characters()
    progressbar.value = 3
    label.value = 'Getting profile data (triumphs)...'
    refresh_triumphs()
    progressbar.value = 4
    label.value = 'Getting profile data (pursuits)...'
    get_pursuits()
    progressbar.value = 5
    refresh_pursuits(progressbar=progressbar)
    progressbar.value = 6
    label.value = 'Complete!'
    get_all_tasks()
    label.value = 'Complete2!'
    filter_tasks()
    label.value = 'Complete3!'
    refresh_task_view()
    label.value = 'Complete4!'
    
    output.clear_output()
    with output: display(interface)
```

```python
all_tasks = {}
def get_all_tasks():
    global all_tasks
    
    for triumph in triumphs:
        all_tasks[triumph] = triumphs[triumph]
        all_tasks[triumph]['type'] = 'Triumph'

    for pursuit in active_pursuits:
        all_tasks[pursuit] = active_pursuits[pursuit]
        all_tasks[pursuit]['type'] = 'Pursuit'

    for pursuit in pursuits_for_sale:
        all_tasks[pursuit] = pursuits_for_sale[pursuit]
        all_tasks[pursuit]['type'] = 'Pursuit'
        all_tasks[pursuit]['complete'] = False
            
```

```python
active_objectives = []
total_potential_tasks = 0
tasks_per_page = 15
total_task_pages = 0
current_page = 0

def filter_tasks():
    
    global potential_tasks
    global total_potential_tasks
    global total_task_pages
    global current_page
    
    current_page = 0
    
    potential_tasks = []

    for task in all_tasks:
        add_task = True
        
        #filter complete
        if task_filter.complete.value and all_tasks[task]['complete']:
            add_task = False
                
        if task_filter.show.value != 'All':
            if task_filter.show.value != 'Triumphs' and all_tasks[task]['type'] == 'Triumph':
                add_task = False
                
            if task_filter.show.value != 'Pursuits' and all_tasks[task]['type'] == 'Pursuit':
                add_task = False
        
        
        if task_filter.conflicting.value:
            conflicting = True
            for objective in all_tasks[task]['objectives']:
                if not has_conflict(objective):
                    conflicting = False
            if conflicting:
                add_task = False
        
        if add_task:
            potential_tasks.append(task) 
 
    total_potential_tasks = len(potential_tasks)      
    total_task_pages = int(total_potential_tasks/tasks_per_page)
    if total_potential_tasks%tasks_per_page > 0:
        total_task_pages += 1

```

```python
def refresh_task_view():
    
    global current_page
    
    if current_page == 0:
        task_filter.previous.disabled = True
    else:
        task_filter.previous.disabled = False
    if current_page == total_task_pages - 1:
        task_filter.next.disabled = True
    else:
        task_filter.next.disabled = False
    
    starting_task = current_page*tasks_per_page
    
    if total_potential_tasks == 0:
        task_filter.label.value = 'No tasks to show!!'
    elif starting_task+tasks_per_page > total_potential_tasks:
        task_filter.label.value = f'Showing Tasks {starting_task+1}-{total_potential_tasks} of {total_potential_tasks}'
    else:
        task_filter.label.value = f'Showing Tasks {starting_task+1}-{starting_task+tasks_per_page} of {total_potential_tasks}'

    for i in range(tasks_per_page):
        
        #hide extra task boxes
        if starting_task + i >= len(all_tasks):
            task_widgets[i].task.layout.display = 'none'
        else:
            task_widgets[i].task.layout.display = 'flex'
            task = potential_tasks[starting_task+i]
            
        
            #update values in task box
            task_path = all_tasks[task]['path']
            task_widgets[i].path[0].value = f'<font size="1">{task_path}</font size=>'

            task_name = all_tasks[task]['name']
            task_widgets[i].name[0].value = f'<font size="5"><b>{task_name}</b></font size=>'

            task_description = all_tasks[task]['description']
            task_widgets[i].description[0].value = f'<i>{task_description}</i>'

            for j in range(6):
                if j >= len(all_tasks[task]['objectives']):
                    task_widgets[i].objectives.children[j].layout.display = 'none'
                else:
                    task_widgets[i].objectives.children[j].layout.display = 'flex'

                    objective_hash = list(all_tasks[task]['objectives'].keys())[j]

                    task_widgets[i].objectives.children[j].children[0].objective_hash = objective_hash
                    task_widgets[i].objectives.children[j].children[3].objective_hash = objective_hash
                    task_widgets[i].objectives.children[j].children[3].task_hash = task

                    objective_completion_value = all_tasks[task]['objectives'][objective_hash]['completionValue']
                    task_widgets[i].objectives.children[j].children[1].max = objective_completion_value

                    objective_progress = all_tasks[task]['objectives'][objective_hash]['progress']
                    task_widgets[i].objectives.children[j].children[1].value = objective_progress

                    objective_description = all_tasks[task]['objectives'][objective_hash]['description']
                    indent = '\xa0\xa0\xa0\xa0'
                    task_widgets[i].objectives.children[j].children[2].value = f'{indent}{objective_description} ({objective_progress}/{objective_completion_value})'
                    
                    #set objective color
                    if objective_progress >= objective_completion_value:
                        task_widgets[i].objectives.children[j].children[0].button_style = 'success'
                        task_widgets[i].objectives.children[j].children[0].disabled = True
                        task_widgets[i].objectives.children[j].children[1].bar_style = 'success'
                    elif objective_hash in active_objectives:
                        task_widgets[i].objectives.children[j].children[0].button_style = 'info'
                        task_widgets[i].objectives.children[j].children[0].disabled = False
                        task_widgets[i].objectives.children[j].children[1].bar_style = 'info'
                    elif has_conflict(objective_hash):
                        task_widgets[i].objectives.children[j].children[0].button_style = 'danger'
                        task_widgets[i].objectives.children[j].children[0].disabled = True
                        task_widgets[i].objectives.children[j].children[1].bar_style = 'danger'
                    else:
                        task_widgets[i].objectives.children[j].children[0].button_style = ''
                        task_widgets[i].objectives.children[j].children[0].disabled = False
                        task_widgets[i].objectives.children[j].children[1].bar_style = ''

```

```python
def has_conflict(objective):
    conflict = False
    if objective in requirements:
        for requirement in requirements[objective]:
            if requirements[objective][requirement] != 'Any':
                if requirement == 'Activity Type' and current_loadout.activity_type.value != requirements[objective][requirement]:
                    if current_loadout.activity_type.value != 'Any':
                        conflict = True
                if requirement == 'Activity' and current_loadout.activity.value != requirements[objective][requirement]:
                    if current_loadout.activity.value != 'Any':
                        conflict = True
                if requirement == 'Class' and current_loadout.c_class.value != requirements[objective][requirement]:
                    if current_loadout.c_class.value != 'Any':
                        conflict = True
                if requirement == 'Subclass' and current_loadout.subclass.value != requirements[objective][requirement]:
                    if current_loadout.subclass.value != 'Any':
                        conflict = True
                if requirement == 'Tree' and current_loadout.tree.value != requirements[objective][requirement]:
                    if current_loadout.tree.value != 'Any':
                        conflict = True
                if requirement == 'Grenade' and current_loadout.grenade.value != requirements[objective][requirement]:
                    if current_loadout.grenade.value != 'Any':
                        conflict = True
    return conflict
```

```python
def set_loadout(dropdown, value):
    dropdown.value = value
    if value == 'Any':
        dropdown.disabled = False
        dropdown.layout = layout_dropdown_inactive
    else:
        dropdown.disabled = True
        dropdown.layout = layout_dropdown_active

def refresh_loadout():
    
    with output: print('test1')
    
    set_loadout(current_loadout.activity_type, 'Any')
    set_loadout(current_loadout.activity, 'Any')
    set_loadout(current_loadout.c_class, 'Any')
    set_loadout(current_loadout.subclass, 'Any')
    set_loadout(current_loadout.tree, 'Any')
    set_loadout(current_loadout.grenade, 'Any')
    set_loadout(current_loadout.kinetic_weapon, 'Any')
    set_loadout(current_loadout.energy_weapon, 'Any')
    set_loadout(current_loadout.energy_type, 'Any')
    set_loadout(current_loadout.power_weapon, 'Any')
    set_loadout(current_loadout.power_type, 'Any')
    set_loadout(current_loadout.armor, 'Any')
    
    with output: print('test2')
    
    for objective in active_objectives:
        for requirement in requirements[objective]:
            if requirements[objective][requirement] != 'Any':
                if requirement == 'Activity Type':
                    set_loadout(current_loadout.activity_type, requirements[objective][requirement])
                if requirement == 'Activity':
                    set_loadout(current_loadout.activity, requirements[objective][requirement])
                if requirement == 'Class':
                    set_loadout(current_loadout.c_class, requirements[objective][requirement])
                if requirement == 'Subclass':
                    set_loadout(current_loadout.subclass, requirements[objective][requirement])
                if requirement == 'Tree':
                    set_loadout(current_loadout.tree, requirements[objective][requirement])
                if requirement == 'Grenade':
                    set_loadout(current_loadout.grenade, requirements[objective][requirement])
                if requirement == 'Weapon Slot':
                    if requirements[objective][requirement] == 'Kinetic':
                        set_loadout(current_loadout.kinetic_weapon, requirements[objective]['Weapon Type'])
                    if requirements[objective][requirement] == 'Energy':
                        set_loadout(current_loadout.energy_weapon, requirements[objective]['Weapon Type'])
                        set_loadout(current_loadout.energy_type, requirements[objective]['Weapon Energy'])
                    if requirements[objective][requirement] == 'Power':
                        set_loadout(current_loadout.power_weapon, requirements[objective]['Weapon Type'])
                        set_loadout(current_loadout.power_type, requirements[objective]['Weapon Energy'])
```
