```python
import ipywidgets as ipw
import pickle

with open('loadout_options.data', 'rb') as filehandle:
    loadout_options = pickle.load(filehandle)
    
```

```python
class Loadout:
    def __init__(self, 
                 activity_type = 'None', 
                 activity = 'None', 
                 c_class = 'None', 
                 subclass = 'None', 
                 tree = 'None', 
                 grenade = 'None', 
                 kinetic_weapon = 'None', 
                 energy_weapon = 'None', 
                 energy_type = 'None', 
                 power_weapon = 'None', 
                 power_type = 'None', 
                 armor = 'None'):
        
        self.activity_type = activity_type
        self.activity = activity
        self.c_class = c_class
        self.subclass = subclass
        self.tree = tree
        self.grenade = grenade
        self.kinetic_weapon = kinetic_weapon
        self.energy_weapon = energy_weapon
        self.energy_type = energy_type
        self.power_weapon = power_weapon
        self.power_type = power_type
        self.armor = armor
        
current_loadout = Loadout()     
```

```python
def build_loadout_panel():
    global current_loadout
    
    #Dropdown layouts ---------------------------
    
    layout_dropdown_inactive = ipw.Layout(
        width='98%',
        border = '')

    layout_dropdown_active = ipw.Layout(
        width='98%',
        border = '2px solid #00bcd4')
    
    #Activity VBox -------------------------------
    
    label_activity_options = ipw.HTML(
        value = "<b>Activity Options</b>")

    
    dropdown_activity_type = ipw.Dropdown(
        options = loadout_options['Activity Type'],
        layout = layout_dropdown_inactive)
    current_loadout.activity_type = dropdown_activity_type

    dropdown_activity = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.activity = dropdown_activity

    vbox_activity = ipw.VBox(
        children=[
            label_activity_options,
            dropdown_activity_type,
            dropdown_activity])
    
    #Class VBox -----------------------------------=
    
    label_class_options = ipw.HTML(
        value = "<b>Class Options</b>")

    label_class = ipw.Label(
        value='Class:')
    
    dropdown_class = ipw.Dropdown(
        options = loadout_options['Class'],
        layout = layout_dropdown_inactive)
    current_loadout.c_class = dropdown_class

    label_subclass = ipw.Label(
        value='Subclass:')
    
    dropdown_subclass = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.subclass = dropdown_subclass

    label_tree = ipw.Label(
        value='Tree:')
    
    dropdown_tree = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.tree = dropdown_tree

    label_grenade = ipw.Label(
        value='Grenade:')
    
    dropdown_grenade = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    current_loadout.grenade = dropdown_grenade

    vbox_class_options = ipw.VBox(
        children = [
            label_class_options,
            label_class,
            dropdown_class,
            label_subclass,
            dropdown_subclass,
            label_tree,
            dropdown_tree,
            label_grenade,
            dropdown_grenade])
    
    
    #Gear VBox -----------------------------------------
    
    label_gear_options = ipw.HTML(
        value = "<b>Gear Options</b>")

    label_kinetic = ipw.Label(
        value='Kinetic Weapon:')
    
    dropdown_kinetic = ipw.Dropdown(
        options = loadout_options['Kinetic Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.kinetic_weapon = dropdown_kinetic

    label_energy = ipw.Label(
        value='Energy Weapon:')
    
    dropdown_energy = ipw.Dropdown(
        options = loadout_options['Energy Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.energy_weapon = dropdown_energy
    
    dropdown_energy_type = ipw.Dropdown(
        options = loadout_options['Energy Weapon Type'],
        layout = layout_dropdown_inactive)
    current_loadout.energy_type = dropdown_energy_type

    label_power = ipw.Label(
        value='Power Weapon:')
    
    dropdown_power = ipw.Dropdown(
        options = loadout_options['Power Weapon'],
        layout = layout_dropdown_inactive)
    current_loadout.power_weapon = dropdown_power
    
    dropdown_power_type = ipw.Dropdown(
        options = loadout_options['Power Weapon Type'],
        layout = layout_dropdown_inactive)
    current_loadout.power_type = dropdown_power_type

    label_armor = ipw.Label(
        value='Armor Set:')
    
    dropdown_armor = ipw.Dropdown(
        options = loadout_options['Armor'],
        layout = layout_dropdown_inactive)
    current_loadout.armor = dropdown_armor
    

    vbox_gear = ipw.VBox(
        children = [
            label_gear_options,
            label_kinetic,
            dropdown_kinetic,
            label_energy,
            dropdown_energy,
            dropdown_energy_type,
            label_power,
            dropdown_power,
            dropdown_power_type,
            label_armor,
            dropdown_armor])
    
    #Loadout Panel -------------------------------
    
    box_spacer = ipw.Box(
        layout=ipw.Layout(
            height='20px'))
    
    vbox_options = ipw.VBox(
        children=[
            vbox_activity,
            box_spacer,
            vbox_class_options,
            box_spacer,
            vbox_gear],
        layout = ipw.Layout(
            border ='1px solid white',
            width ='15%',
            min_width = '250px',
            padding ='10px'))


    #Interact Functions ----------------------------

    def change_class(change):
        if change['new'] == 'Any':
            dropdown_subclass.options = ['Any']
            dropdown_subclass.disabled = True
        else:
            dropdown_subclass.options = loadout_options['Subclass'][change['new']]
            dropdown_subclass.disabled = False

    def change_subclass(change):
        if change['new'] == 'Any':
            dropdown_tree.options = ['Any']
            dropdown_tree.disabled = True
            dropdown_grenade.options = ['Any']
            dropdown_grenade.disabled = True
        else:
            dropdown_tree.options = loadout_options['Tree'][dropdown_class.value][change['new']]
            dropdown_tree.disabled = False
            dropdown_grenade.options = loadout_options['Grenade'][dropdown_class.value][change['new']]
            dropdown_grenade.disabled = False

    def change_activity_type(change):
        if change['new'] == 'Any':
            dropdown_activity.options = ['Any']
            dropdown_activity.disabled = True
        else:
            dropdown_activity.options = loadout_options['Activity'][change['new']]
            dropdown_activity.disabled = False

    dropdown_class.observe(change_class, names = 'value')
    dropdown_subclass.observe(change_subclass, names = 'value')
    dropdown_activity_type.observe(change_activity_type, names = 'value')
    
    return vbox_options
```

```python
class Task:
    def __init__(self, path, name, description, objectives):
        self.path = path,
        self.name = name,
        self.description = description,
        self.objectives = objectives
        
        
```

```python
def build_objective():
        
        button_objective = ipw.Button(
            layout = ipw.Layout(
                width = '32px',
                height = '32px',
                order = '-2'))
        #button_objective.on_click(toggle_objective)
        
        intProgress_objective = ipw.IntProgress(
            value=30,
            min=0,
            max=100,
            step=1,
            layout = ipw.Layout(
                width = '100%',
                height = '32px',
                overflow = 'visible'))
        
        label_objective = ipw.Label(
             value = f'\xa0\xa0\xa0objective', 
             layout = ipw.Layout(
                padding='1px',
                width = '0px',
                overflow = 'visible',
                order='-1'))
            
        button_edit = ipw.Button(
            description ='EDIT', 
            layout = ipw.Layout(
                width='60px'))
        #button_edit.on_click(edit_objective)
        
        hbox_objective = ipw.HBox(
            children = [
                button_objective,
                intProgress_objective, 
                label_objective, 
                button_edit])
        
        return hbox_objective
```

```python
task_widgets = []

def build_task_widget():
    global task_widgets

    label_path = ipw.HTML(
        value = '<font size="1">Path/Path/Path/Path/</font size>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'))
    
    label_name = ipw.HTML(
        value = '<b>Task Name</b>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'))
        
    label_description = ipw.HTML(
        value = '<i>test test test test test test test test test test test test test test test test test test test test test test test test </i>',
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap'))
        
    vbox_objectives = ipw.VBox(
        children = [build_objective() for i in range(6)])

    vbox_triumph = ipw.VBox(
        children = [
            label_path,
            label_name,
            label_description, 
            vbox_objectives],
        layout = ipw.Layout(
            margin = '10px',
            padding='10px',
            border='3px solid white',
            flex='1 1 0%', 
            width='auto',
            min_width='400px',
            overflow = 'visible'))
    
    task = Task(label_path, label_name, label_description, vbox_objectives)
    task_widgets.append(task)

    return vbox_triumph
```

```python
def build_task_panel():
    
    box_tasks = ipw.Box(
        children = [build_task_widget() for i in range(15)], 
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap',
            align_items='stretch'))
    
    return box_tasks
```

```python
def build_filter_panel():
    
    checkbox_complete = ipw.Checkbox(
        description='Hide Complete Objectives')
    
    checkbox_conflicting = ipw.Checkbox(
        description='Hide Conflicting Objectives')
    
    dropdown_show = ipw.Dropdown(
        description = "Show:",
        options= [
            "All",
            "Active Objectives",
            "Pursuits",
            "Milestones",
            "Triumphs",
            "Collections"])
    
    vbox_filters = ipw.VBox(
        children = [
            dropdown_show,
            checkbox_complete,
            checkbox_conflicting])
    
    label_current_page = ipw.HTML(
        value = 'Showing __-__ of ___')
    
    button_previous_page = ipw.Button(
        description = 'Previous',
        layout = ipw.Layout(
            width='100px'))
    
    button_next_page = ipw.Button(
        description = 'Next',
        layout = ipw.Layout(
            width='100px'))
    
    hbox_change_page = ipw.HBox(
        children = [
            button_previous_page,
            button_next_page])
    
    button_refresh = ipw.Button(
        description='Refresh',
        layout = ipw.Layout(
            width='186px'))
    
    vbox_page = ipw.VBox(
        children=[
            label_current_page,
            hbox_change_page,
            button_refresh],
        layout = ipw.Layout(
            width='200px',
            padding = '5px',
            justify_items = 'flex-end'))
    
    hbox_filter_panel = ipw.HBox(
        children = [
            vbox_filters,
            vbox_page],
        layout = ipw.Layout(
            justify_content = 'space-between',
            height = '300px'))
    
    
    return hbox_filter_panel
```
