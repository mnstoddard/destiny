```python
all_tasks = {}
filtered_tasks = []
active_objectives = []
v_index = 0

with open('requirements_tasks.data', 'rb') as filehandle:
    requirements = pickle.load(filehandle)

#TODO: Order Tasks
```

```python
def get_all_tasks():
    global all_tasks
    
    for triumph in triumphs:
        all_tasks[triumph] = triumphs[triumph]
        all_tasks[triumph]['type'] = 'Triumph'

    for pursuit in active_pursuits:
        all_tasks[pursuit] = active_pursuits[pursuit]
        all_tasks[pursuit]['type'] = 'Pursuit'

    for pursuit in pursuits_for_sale:
        all_tasks[pursuit] = pursuits_for_sale[pursuit]
        all_tasks[pursuit]['type'] = 'Pursuit'
        all_tasks[pursuit]['complete'] = False
```

```python
def filter_tasks():
    global filtered_tasks
    global v_index
    
    old_filtered_tasks = filtered_tasks
    
    filtered_tasks = []

    for task in all_tasks:
        add_task = True
        
        #filter complete
        if filters.complete.value and all_tasks[task]['complete']:
            add_task = False
        
        #filter conflicting
        if filters.conflicting.value:
            conflicting = True
            for objective in all_tasks[task]['objectives']:
                if not has_conflict(objective):
                    conflicting = False
            if conflicting:
                add_task = False
                
        #filter by path
        filter_str = ''
        for i in range(len(filters.path_filters)):
            if filters.path_filters[i].value != 'All':
                if filter_str == '':
                    filter_str += filters.path_filters[i].value
                else: 
                    filter_str += '/' + filters.path_filters[i].value
        if filter_str not in all_tasks[task]['path']:
                add_task = False
        
        if add_task:
            filtered_tasks.append(task) 
            
    for i in range(v_index, len(old_filtered_tasks)):
        if old_filtered_tasks[i] in filtered_tasks:
            v_index = filtered_tasks.index(old_filtered_tasks[i])
            break;
            
    update_path_filters()
    refresh_view()
 
```

```python
def set_page_info():
    global v_index
    
    n_tasks = len(filtered_tasks)
    n_pages = n_tasks/tasks_per_page
    if n_pages%1>0:
        n_pages = int(n_pages) + 1
    else:
        n_pages = int(n_pages)
    
    if v_index == 0:
        filters.previous.disabled = True
    else:
        filters.previous.disabled = False
    if v_index + tasks_per_page >= n_tasks:
        filters.next.disabled = True
    else:
        filters.next.disabled = False

    
    if n_tasks == 0:
        filters.label.value = 'No tasks to show!!'
    elif v_index+tasks_per_page > n_tasks:
        filters.label.value = f'Showing Tasks {v_index+1}-{n_tasks} of {n_tasks}'
    else:
        filters.label.value = f'Showing Tasks {v_index+1}-{v_index+tasks_per_page} of {n_tasks}'
    
    
    
```

```python
def refresh_view():
    global tasks
    
    set_page_info()
    
    for i in range(tasks_per_page):
        
        #hide extra task boxes
        if v_index + i >= len(filtered_tasks):
            tasks.list[i].task.layout.display = 'none'
        else:
            tasks.list[i].task.layout.display = 'flex'
            task = filtered_tasks[v_index + i]
            
            #update values in task box
            task_path = all_tasks[task]['path']
            tasks.list[i].path[0].value = f'<font size="1">{task_path}</font size=>'

            task_name = all_tasks[task]['name']
            tasks.list[i].name[0].value = f'<font size="5"><b>{task_name}</b></font size=>'

            task_description = all_tasks[task]['description']
            tasks.list[i].description[0].value = f'<i>{task_description}</i>'

            for j in range(6):
                objective = tasks.list[i].objectives.children[j]
                
                if j >= len(all_tasks[task]['objectives']):
                    objective.layout.display = 'none'
                else:
                    objective.layout.display = 'flex'

                    objective_hash = list(all_tasks[task]['objectives'].keys())[j]

                    objective.children[0].objective_hash = objective_hash
                    objective.children[3].objective_hash = objective_hash
                    objective.children[3].task_hash = task

                    objective_completion_value = all_tasks[task]['objectives'][objective_hash]['completionValue']
                    objective.children[1].max = objective_completion_value

                    objective_progress = all_tasks[task]['objectives'][objective_hash]['progress']
                    objective.children[1].value = objective_progress

                    objective_description = all_tasks[task]['objectives'][objective_hash]['description']
                    indent = '\xa0\xa0\xa0\xa0'
                    objective.children[2].value = f'{indent}{objective_description} ({objective_progress}/{objective_completion_value})'
                    
                    #set objective color
                    if objective_progress >= objective_completion_value:
                        objective.children[0].button_style = 'success'
                        objective.children[0].disabled = True
                        objective.children[1].bar_style = 'success'
                    elif objective_hash in active_objectives:
                        objective.children[0].button_style = 'info'
                        objective.children[0].disabled = False
                        objective.children[1].bar_style = 'info'
                    elif has_conflict(objective_hash):
                        objective.children[0].button_style = 'danger'
                        objective.children[0].disabled = True
                        objective.children[1].bar_style = 'danger'
                    else:
                        objective.children[0].button_style = ''
                        objective.children[0].disabled = False
                        objective.children[1].bar_style = ''

```

```python
def has_conflict(objective):
    conflict = False
    if objective in requirements:
        for requirement in requirements[objective]:
            if requirements[objective][requirement] != 'Any':
                if requirement == 'Activity Type' and loadout.activity_type.value != requirements[objective][requirement]:
                    if loadout.activity_type.value != 'Any':
                        conflict = True
                if requirement == 'Activity' and loadout.activity.value != requirements[objective][requirement]:
                    if loadout.activity.value != 'Any':
                        conflict = True
                if requirement == 'Class' and loadout.c_class.value != requirements[objective][requirement]:
                    if loadout.c_class.value != 'Any':
                        conflict = True
                if requirement == 'Subclass' and loadout.subclass.value != requirements[objective][requirement]:
                    if loadout.subclass.value != 'Any':
                        conflict = True
                if requirement == 'Tree' and loadout.tree.value != requirements[objective][requirement]:
                    if loadout.tree.value != 'Any':
                        conflict = True
                if requirement == 'Grenade' and loadout.grenade.value != requirements[objective][requirement]:
                    if loadout.grenade.value != 'Any':
                        conflict = True
    return conflict
```

```python
def refresh_loadout():
    
    def set_loadout(dropdown, value):
        dropdown.value = value
        if value == 'Any':
            dropdown.disabled = False
            dropdown.layout = loadout.layout_dropdown_inactive
        else:
            dropdown.disabled = True
            dropdown.layout = loadout.layout_dropdown_active
    
    set_loadout(loadout.activity_type, 'Any')
    set_loadout(loadout.activity, 'Any')
    set_loadout(loadout.c_class, 'Any')
    set_loadout(loadout.subclass, 'Any')
    set_loadout(loadout.tree, 'Any')
    set_loadout(loadout.grenade, 'Any')
    set_loadout(loadout.kinetic_weapon, 'Any')
    set_loadout(loadout.energy_weapon, 'Any')
    set_loadout(loadout.energy_type, 'Any')
    set_loadout(loadout.power_weapon, 'Any')
    set_loadout(loadout.power_type, 'Any')
    set_loadout(loadout.armor, 'Any')
    
    for objective in active_objectives:
        for requirement in requirements[objective]:
            if requirements[objective][requirement] != 'Any':
                if requirement == 'Activity Type':
                    set_loadout(loadout.activity_type, requirements[objective][requirement])
                if requirement == 'Activity':
                    set_loadout(loadout.activity, requirements[objective][requirement])
                if requirement == 'Class':
                    set_loadout(loadout.c_class, requirements[objective][requirement])
                if requirement == 'Subclass':
                    set_loadout(loadout.subclass, requirements[objective][requirement])
                if requirement == 'Tree':
                    set_loadout(loadout.tree, requirements[objective][requirement])
                if requirement == 'Grenade':
                    set_loadout(loadout.grenade, requirements[objective][requirement])
                if requirement == 'Weapon Slot':
                    if requirements[objective][requirement] == 'Kinetic':
                        set_loadout(loadout.kinetic_weapon, requirements[objective]['Weapon Type'])
                    if requirements[objective][requirement] == 'Energy':
                        set_loadout(loadout.energy_weapon, requirements[objective]['Weapon Type'])
                        set_loadout(loadout.energy_type, requirements[objective]['Weapon Energy'])
                    if requirements[objective][requirement] == 'Power':
                        set_loadout(loadout.power_weapon, requirements[objective]['Weapon Type'])
                        set_loadout(loadout.power_type, requirements[objective]['Weapon Energy'])
```

```python
def update_path_filters():

    path_map = {}

    def add_path_level(current_path, remaining_branches):
        if remaining_branches[0] not in current_path:
            current_path[remaining_branches[0]] = {}
        if len(remaining_branches) > 1:
            add_path_level(current_path[remaining_branches[0]], remaining_branches[1:])

    for task in filtered_tasks: 
        add_path_level(path_map, all_tasks[task]['path'].split('/'))
        
    for i in range(len(filters.path_filters)):
        filters.path_filters[i].unobserve(modify_filter, 'value')
    
    def get_options(i):
        temp_dict = path_map
        for j in range(i):
            temp_dict = temp_dict[filters.path_filters[j].value]
        return ['All'] + list(temp_dict.keys())

    old_value = filters.path_filters[0].value
    options = get_options(0)
    filters.path_filters[0].options = options
    if old_value in options:
        filters.path_filters[0].value = old_value
    
    for i in range(1,len(filters.path_filters)):
        if filters.path_filters[i-1].value == 'All':
            filters.path_filters[i].layout.display = 'none'
            filters.path_filters[i].value = 'All'
            filters.path_filters[i].options = ['All']
        else:
            old_value = filters.path_filters[i].value
            options = get_options(i)
            filters.path_filters[i].options = options
            if old_value in options:
                filters.path_filters[i].value = old_value
            filters.path_filters[i].layout.display = 'flex'
            if len(filters.path_filters[i].options) == 1:
                filters.path_filters[i].layout.display = 'none'
                
        
        for i in range(len(filters.path_filters)):
            filters.path_filters[i].observe(modify_filter, 'value')
        
```
