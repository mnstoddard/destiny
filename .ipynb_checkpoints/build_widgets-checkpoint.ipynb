```python
import ipywidgets as ipw
import pickle

with open('loadout_options.data', 'rb') as filehandle:
    loadout_options = pickle.load(filehandle)
    
tasks_per_page = 15
```

```python
class global_widget:
    def __init__(self):
        self
        
class Task:
    def __init__(self, task, path, name, description, objectives):
        self.task = task
        self.path = path,
        self.name = name,
        self.description = description,
        self.objectives = objectives
        
   
        
loadout = global_widget()
tasks = global_widget()
filters = global_widget()
edit = global_widget()
%run edit_panel.ipynb
```

```python
def build_loadout_panel():
    global loadout
    
    #Dropdown layouts ---------------------------
    layout_dropdown_inactive = ipw.Layout(
        width='98%',
        border = '')
    loadout.layout_dropdown_inactive = layout_dropdown_inactive
    
    layout_dropdown_active = ipw.Layout(
        width='98%',
        border = '2px solid #00bcd4')
    loadout.layout_dropdown_active = layout_dropdown_active
    
    #Activity VBox -------------------------------
    label_activity_options = ipw.HTML(
        value = "<b>Activity Options</b>")

    
    dropdown_activity_type = ipw.Dropdown(
        options = loadout_options['Activity Type'],
        layout = layout_dropdown_inactive)
    loadout.activity_type = dropdown_activity_type

    dropdown_activity = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    loadout.activity = dropdown_activity

    vbox_activity = ipw.VBox(
        children=[
            label_activity_options,
            dropdown_activity_type,
            dropdown_activity])
    
    #Class VBox -----------------------------------=
    
    label_class_options = ipw.HTML(
        value = "<b>Class Options</b>")

    label_class = ipw.Label(
        value='Class:')
    
    dropdown_class = ipw.Dropdown(
        options = loadout_options['Class'],
        layout = layout_dropdown_inactive)
    loadout.c_class = dropdown_class

    label_subclass = ipw.Label(
        value='Subclass:')
    
    dropdown_subclass = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    loadout.subclass = dropdown_subclass

    label_tree = ipw.Label(
        value='Tree:')
    
    dropdown_tree = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    loadout.tree = dropdown_tree

    label_grenade = ipw.Label(
        value='Grenade:')
    
    dropdown_grenade = ipw.Dropdown(
        options = ['Any'],
        disabled = True,
        layout = layout_dropdown_inactive)
    loadout.grenade = dropdown_grenade

    vbox_class_options = ipw.VBox(
        children = [
            label_class_options,
            label_class,
            dropdown_class,
            label_subclass,
            dropdown_subclass,
            label_tree,
            dropdown_tree,
            label_grenade,
            dropdown_grenade])
    
    
    #Gear VBox -----------------------------------------
    
    label_gear_options = ipw.HTML(
        value = "<b>Gear Options</b>")

    label_kinetic = ipw.Label(
        value='Kinetic Weapon:')
    
    dropdown_kinetic = ipw.Dropdown(
        options = loadout_options['Kinetic Weapon'],
        layout = layout_dropdown_inactive)
    loadout.kinetic_weapon = dropdown_kinetic

    label_energy = ipw.Label(
        value='Energy Weapon:')
    
    dropdown_energy = ipw.Dropdown(
        options = loadout_options['Energy Weapon'],
        layout = layout_dropdown_inactive)
    loadout.energy_weapon = dropdown_energy
    
    dropdown_energy_type = ipw.Dropdown(
        options = loadout_options['Energy Weapon Type'],
        layout = layout_dropdown_inactive)
    loadout.energy_type = dropdown_energy_type

    label_power = ipw.Label(
        value='Power Weapon:')
    
    dropdown_power = ipw.Dropdown(
        options = loadout_options['Power Weapon'],
        layout = layout_dropdown_inactive)
    loadout.power_weapon = dropdown_power
    
    dropdown_power_type = ipw.Dropdown(
        options = loadout_options['Power Weapon Type'],
        layout = layout_dropdown_inactive)
    loadout.power_type = dropdown_power_type

    label_armor = ipw.Label(
        value='Armor Set:')
    
    dropdown_armor = ipw.Dropdown(
        options = loadout_options['Armor'],
        layout = layout_dropdown_inactive)
    loadout.armor = dropdown_armor
    

    vbox_gear = ipw.VBox(
        children = [
            label_gear_options,
            label_kinetic,
            dropdown_kinetic,
            label_energy,
            dropdown_energy,
            dropdown_energy_type,
            label_power,
            dropdown_power,
            dropdown_power_type,
            label_armor,
            dropdown_armor])
    
    #Loadout Panel -------------------------------
    
    box_spacer = ipw.Box(
        layout=ipw.Layout(
            height='20px'))
    
    vbox_options = ipw.VBox(
        children=[
            vbox_activity,
            box_spacer,
            vbox_class_options,
            box_spacer,
            vbox_gear],
        layout = ipw.Layout(
            border ='1px solid white',
            width ='15%',
            min_width = '250px',
            padding ='10px'))


    #Interact Functions ----------------------------

    def change_class(change):
        if change['new'] == 'Any':
            dropdown_subclass.options = ['Any']
            dropdown_subclass.disabled = True
        else:
            dropdown_subclass.options = loadout_options['Subclass'][change['new']]
            dropdown_subclass.disabled = False

    def change_subclass(change):
        if change['new'] == 'Any':
            dropdown_tree.options = ['Any']
            dropdown_tree.disabled = True
            dropdown_grenade.options = ['Any']
            dropdown_grenade.disabled = True
        else:
            dropdown_tree.options = loadout_options['Tree'][dropdown_class.value][change['new']]
            dropdown_tree.disabled = False
            dropdown_grenade.options = loadout_options['Grenade'][dropdown_class.value][change['new']]
            dropdown_grenade.disabled = False

    def change_activity_type(change):
        if change['new'] == 'Any':
            dropdown_activity.options = ['Any']
            dropdown_activity.disabled = True
        else:
            dropdown_activity.options = loadout_options['Activity'][change['new']]
            dropdown_activity.disabled = False

    dropdown_class.observe(change_class, names = 'value')
    dropdown_subclass.observe(change_subclass, names = 'value')
    dropdown_activity_type.observe(change_activity_type, names = 'value')
    
    loadout.panel = vbox_options
```

```python
def build_objective():
        
    button_objective = ipw.Button(
        layout = ipw.Layout(
            width = '32px',
            height = '32px',
            order = '-2'
        )
    )

    intProgress_objective = ipw.IntProgress(
        value=30,
        min=0,
        max=100,
        step=1,
        layout = ipw.Layout(
            width = '100%',
            height = '32px',
            overflow = 'visible'
        )
    )

    label_objective = ipw.Label(
         value = f'\xa0\xa0\xa0objective', 
         layout = ipw.Layout(
            padding='1px',
            width = '0px',
            overflow = 'visible',
            order='-1'))

    button_edit = ipw.Button(
        description ='EDIT', 
        layout = ipw.Layout(
            width='60px',
            display = 'none'
        )
    )
    

    hbox_objective = ipw.HBox(
        children = [
            button_objective,
            intProgress_objective, 
            label_objective, 
            button_edit
        ]
    )
        
    def toggle_objective(b):
        global active_objectives
        if b.objective_hash in active_objectives:
            active_objectives.remove(b.objective_hash)
        else:
            active_objectives.append(b.objective_hash)
        refresh_loadout()
        filter_tasks()
    button_objective.on_click(toggle_objective)
    
    
    def edit_objective(b):
        dropdown_triumph_req.value = b.task_hash
        dropdown_objective_req.value = b.objective_hash
    button_edit.on_click(edit_objective)
        
    return hbox_objective
```

```python
tasks.list = []

def build_task_widget():

    label_path = ipw.HTML(
        value = '<font size="1">Path/Path/Path/Path/</font size>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'
        )
    )
    
    label_name = ipw.HTML(
        value = '<b>Task Name</b>', 
        layout = ipw.Layout(
            width = '100%',
            overflow = 'visible'
        )
    )
        
    label_description = ipw.HTML(
        value = '<i>test test test test test test test test </i>',
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap'
        )
    )
        
    vbox_objectives = ipw.VBox(
        children = [build_objective() for i in range(6)]
    )

    vbox_triumph = ipw.VBox(
        children = [
            label_path,
            label_name,
            label_description, 
            vbox_objectives
        ],
        layout = ipw.Layout(
            margin = '10px',
            padding='10px',
            border='3px solid white',
            flex='1 1 0%', 
            width='auto',
            min_width='400px',
            overflow = 'visible'
        )
    )
    
    task = Task(vbox_triumph, label_path, label_name, label_description, vbox_objectives)
    tasks.list.append(task)
    
    return vbox_triumph
```

```python
def build_task_panel():
    
    box_tasks = ipw.Box(
        children = [build_task_widget() for i in range(tasks_per_page)], 
        layout = ipw.Layout(
            display = 'flex',
            flex_flow = 'row wrap',
            align_items='stretch'
        )
    )
    
    tasks.panel = box_tasks
```

```python
def modify_filter(b):
    filter_tasks()

def build_filter_panel():

    #path Filters
    path_filters = []

    for i in range(10):
        path_filter = ipw.Dropdown(
            options = ['All'],
            layout = ipw.Layout(
                width = '95%'
            )
        )
        path_filter.observe(modify_filter, 'value')
        path_filters.append(path_filter)
    
    filters.path_filters = path_filters
    
    label_filter = ipw.Label(value = 'Filter Tasks')
    
    vbox_path = ipw.VBox(
        children = path_filters,
        layout = ipw.Layout(width = '100%')
    )
    
    #main filters -----------------------------
    checkbox_complete = ipw.Checkbox(
        description = 'Hide Complete Tasks',
        indent = False
    )
    filters.complete = checkbox_complete
    checkbox_complete.observe(modify_filter)
    
    checkbox_conflicting = ipw.Checkbox(
        description = 'Hide Conflicting Tasks',
        indent = False
    )
    filters.conflicting = checkbox_conflicting
    checkbox_conflicting.observe(modify_filter)
    
    checkbox_expired = ipw.Checkbox(
        description = 'Hide Expired Tasks',
        indent = False
    )
    filters.expired = checkbox_expired
    checkbox_expired.observe(modify_filter)
    
    vbox_filters = ipw.VBox(
        children = [
            checkbox_complete,
            checkbox_conflicting,
            checkbox_expired
        ],
        layout = ipw.Layout(
            
        )
    )
    
    label_current_page = ipw.HTML(
        value = 'Showing __-__ of ___'
    )
    filters.label = label_current_page
    
    button_previous_page = ipw.Button(
        description = 'Previous',
        layout = ipw.Layout(
            width='100px'
        )
    )
    filters.previous = button_previous_page
    
    button_next_page = ipw.Button(
        description = 'Next',
        layout = ipw.Layout(
            width='100px'
        )
    )
    filters.next = button_next_page
    
    hbox_change_page = ipw.HBox(
        children = [
            button_previous_page,
            button_next_page
        ]
    )
    
    button_refresh = ipw.Button(
        description='Refresh',
        layout = ipw.Layout(
            width='186px'
        )
    )
    filters.refresh = button_refresh
    
    vbox_page = ipw.VBox(
        children=[
            label_current_page,
            hbox_change_page,
            button_refresh],
        layout = ipw.Layout(
            width='200px',
            padding = '5px',
            justify_items = 'center'))
    
    vbox_filter_panel = ipw.VBox(
        children = [
            vbox_page,
            vbox_filters,
            vbox_path,
        ],
        layout = ipw.Layout(
            border ='1px solid white',
            width ='15%',
            min_width = '250px',
            padding ='10px'
        )
    )
    
    def previous_page(b):
        global v_index 
        v_index += -tasks_per_page
        if v_index <= 0:
            v_index = 0
        refresh_view()


    def next_page(b):
        global v_index 
        v_index += tasks_per_page
        if v_index + tasks_per_page > len(filtered_tasks):
            v_index = len(filtered_tasks) - tasks_per_page
        refresh_view()

    button_previous_page.on_click(previous_page)
    button_next_page.on_click(next_page)


    def refresh_button(b):
        filter_tasks()
        refresh_view()

    button_refresh.on_click(refresh_button)
    
    
    filters.panel = vbox_filter_panel
```
